```python
def signedsquareroot(x):
    x = np.array(x)
    positive_root = np.sqrt(x)
    negative_root = -np.sqrt(x)
    z = positive_root
    for i in range(len(z)-1):
        if z[i+1]-z[i]>0:
            z[i:] = negative_root[i:]
            break
    return z
```

```python
def findtheroot(abst, x):
    x = np.array(x)
    y = x[1:]
    z = x[:-1]
    index = np.argwhere(y*z<0)[0][0]
    y1 = x[index]
    y2 = x[index+1]
    return abst[index] + (abst[index+1] - abst[index])/(y1-y2) * y1
```
